```python
# =========================
# 1. LIBRERÍAS
# =========================
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.feature_extraction.text import CountVectorizer

# =========================
# 2. DATASET (EJEMPLO SIMPLE)
# =========================
data = pd.DataFrame({
    "mensaje": [
        "gana dinero rapido",
        "oferta exclusiva hoy",
        "hola como estas",
        "nos vemos mañana",
        "gana premios gratis",
        "reunion de trabajo",
        "oferta limitada compra ya",
        "te llamo luego"
    ],
    "label": ["spam", "spam", "no spam", "no spam", "spam", "no spam", "spam", "no spam"]
})

# =========================
# 3. VARIABLES
# =========================
X = data["mensaje"]
y = data["label"]

# =========================
# 4. TRANSFORMACIÓN DE TEXTO
# =========================
vectorizer = CountVectorizer()
X_vect = vectorizer.fit_transform(X)

# =========================
# 5. TRAIN / TEST
# =========================
X_train, X_test, y_train, y_test = train_test_split(
    X_vect, y, test_size=0.3, random_state=42
)

# =========================
# 6. MODELO: BAYES INGENUO
# =========================
model = MultinomialNB()
model.fit(X_train, y_train)

# =========================
# 7. PREDICCIÓN
# =========================
y_pred = model.predict(X_test)

# =========================
# 8. MATRIZ DE CONFUSIÓN
# =========================
cm = confusion_matrix(y_test, y_pred)
print("Matriz de confusión:")
print(cm)

print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred))

# =========================
# 9. PRUEBA MANUAL
# =========================
nuevo_mensaje = ["gana dinero gratis ahora"]
nuevo_vect = vectorizer.transform(nuevo_mensaje)

prediccion = model.predict(nuevo_vect)
print("\nMensaje:", nuevo_mensaje[0])
print("Clasificación:", prediccion[0])
```
